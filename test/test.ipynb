```python
import cv2
import matplotlib.pyplot as plt

# c nh gc
img_path = '1004.jpg'
original_image = cv2.imread(img_path, cv2.IMREAD_COLOR)
original_height, original_width = original_image.shape[:2]

# Thay i kch thc nh thnh 256x256
resized_image = cv2.resize(original_image, (256, 256))

# Chuyn nh v kch thc ban u
restored_image = cv2.resize(resized_image, (original_width, original_height))

# Hin th cc nh
plt.figure(figsize=(15, 5))

# nh gc
plt.subplot(1, 3, 1)
plt.title('nh gc')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# nh sau khi resize v 256x256
plt.subplot(1, 3, 2)
plt.title('nh  resize thnh 256x256')
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# nh sau khi khi phc kch thc ban u
plt.subplot(1, 3, 3)
plt.title('nh  khi phc v kch thc ban u')
plt.imshow(cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Đọc ảnh gốc
img_path = '1004.jpg'
original_image = cv2.imread(img_path, cv2.IMREAD_COLOR)
original_height, original_width = original_image.shape[:2]

# Thay đổi kích thước ảnh thành 256x256
resized_image = cv2.resize(original_image, (256, 256))

# Chuyển ảnh về kích thước ban đầu
restored_image = cv2.resize(resized_image, (original_width, original_height))

# Hiển thị các ảnh
plt.figure(figsize=(15, 5))

# Ảnh gốc
plt.subplot(1, 3, 1)
plt.title('Ảnh gốc')
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Ảnh sau khi resize về 256x256
plt.subplot(1, 3, 2)
plt.title('Ảnh đã resize thành 256x256')
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Ảnh sau khi khôi phục kích thước ban đầu
plt.subplot(1, 3, 3)
plt.title('Ảnh đã khôi phục về kích thước ban đầu')
plt.imshow(cv2.cvtColor(restored_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
